```python
def get_content_value(row_data):
    if row_data.find("li"):
        return [li.get_text(" ",strip=True).replace("\xa0"," ") for li in row_data.find_all("li")]
    elif row_data.find("br"):
        return [text for text in row_data.stripped_strings]
    else:
        return row_data.get_text(" ",strip=True).replace("\xa0"," ")

def clean_ref(soup):
    for tag in soup.find_all(["sup","span"]):
        tag.decompose()
        
    
def infobox(url):
    r=requests.get(url)
    soup=bs(r.text)
    
    info_box=soup.find(class_="infobox vevent")
    info_rows=info_box.find_all("tr")
    
    clean_ref(soup)
    movie_info={}    
    for index,row in enumerate(info_rows):
        try:
            if index==0:
                movie_info["title"]=row.find("th").get_text(" ",strip=True)

            else:
                header=row.find('th')
                data=row.find('td')
                if header and data:                
                    content_key=row.find("th").get_text(" ",strip=True)
                    content_value= get_content_value(row.find("td"))
                    movie_info[content_key]=content_value
        except Exception as e:
            print(index,row)
    return movie_info
```

```python
basepath="https://en.wikipedia.org/"
r=requests.get("https://en.wikipedia.org/wiki/List_of_Walt_Disney_Pictures_films")
soup=bs(r.content)
movies=soup.select(".wikitable.sortable i a")
movielist=[]
for index,movie in enumerate(movies):
    try:
        relative_path=movie['href']
        title=movie['title']
        fullpath=basepath+relative_path
        movielist.append(infobox(fullpath))
    except Exception as e:
        print(movie)
        print(e)
```

```python
def min_to_int(runningtime):
    if runningtime=="NA":
        return None
    
    if isinstance(runningtime, list):
        return int(runningtime[0].split(" ")[0])
    else:# is a string
        if "–" in runningtime:
            return int(runningtime.split(" ")[0].split("–")[0])
        else:
            return int(runningtime.split(" ")[0])
    
for movie in movie_info:
    movie['Running time (int)']=min_to_int(movie.get('Running time','NA'))
```

```python
import re
```

```python
amounts=r"thousand|million|billion"
number=r"\d+(,\d{3})*\.*\d*"

word_re= rf"\${number}(–|\sto\s)?({number})?\s({amounts})"
value_re=rf"\${number}"
```

```python
def word_to_value(word):
    value_dict={"thousand":1000,"million":1000000,"billion":1000000000}
    return value_dict[word]

def parse_word_syntax(string):
    value_string=re.search(number,string).group()
    value=float(value_string.replace(",",""))
    word=re.search(amounts,string, flags=re.I).group().lower()
    word_value=word_to_value(word)
    return (value*word_value)
```

```python
def parse_value_syntax(string):
    value_string=re.search(number,string).group()
    value=float(value_string.replace(",",""))
    return value
```

```python
def money_conversion(money):
    
    if money=="NA":
        return None
    if isinstance(money,list):
        money=money[0]
    
    word_syntax= re.search(word_re,money, flags=re.I)
    value_syntax=re.search(value_re,money)
    
    if word_syntax:
        return parse_word_syntax(word_syntax.group())
    elif value_syntax:
        return parse_value_syntax(value_syntax.group())
    else:
        return None
```

```python
print(money_conversion("NA"))
```

```python
for movie in movie_info:
    movie['Box office (int)']=money_conversion(movie.get('Box office','NA'))
    movie['Budget (int)']=money_conversion(movie.get('Budget','NA'))
```

```python
from datetime import datetime
```

```python
dates=[movie.get("Release date","NA") for movie in movie_info]

def cleandate(date):
    return date.split("(")[0].strip()

def date_conversion(date):
    if isinstance(date,list):
        date=date[0]
    
    if date=="NA":
        return None
    
    date_str= cleandate(date)
    fmts=["%B %d, %Y", "%d %B %Y"]
    for fmt in fmts:
        try:
            return datetime.strptime(date_str,fmt)
        except:
            pass
    return None
```

```python
for date in dates: # checking if its working ok
    print(date_conversion(date))
    print()
```

```python
for movie in movie_info:
    movie['Release date (datetime)']=date_conversion(movie.get('Release date','NA'))
```

```python
import requests
import os
import urllib

def get_omdb_info(title):
    base_url="http://www.omdbapi.com/?"
    parameters={"apikey": os.environ['OMDB_API_KEY'], 't': title}
    params_encoded=urllib.parse.urlencode(parameters)
    full_url=base_url+params_encoded
    requests.get(full_url).json()
```

```python
def get_rotten_tomator_score(omdb_info):
    ratings=omdb_info.get('Ratings',[])
    for rating in ratings:
        print(rating)
```

```python
for movie in movie_info:
    title=movie['title']
    omdb_info=get_omdb_info(title)
    movie['imdb']=omdb_info.get('imdbRating',None)
    movie['metascore']=omdb_info.get("Metascore",None)
    movie['rotten_tomatoes']=get_rotten_tomator_score(omdb_info)
```
